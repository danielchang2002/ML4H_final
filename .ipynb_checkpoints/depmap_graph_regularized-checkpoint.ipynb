```python
def evaluate_Challenge_B(df_true, df_score, method='pearson'):
    score_list = []
    df_true = df_true.values
    df_score = df_score.values
    # for col in df_score.columns:
    for i in range(df_score.shape[1]):
        y_pred = df_score[:, i]
        y_true = df_true[:, i]
        # score_list.append(y_true.corr(y_pred, method=method))
        score_list.append(np.corrcoef(y_true, y_pred)[0, 1])
        # score_list.append(accuracy_score(y_pred, y_true))
    return score_list
```
